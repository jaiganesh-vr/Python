```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

fig, ax = plt.subplots()
(line,) = ax.plot([], [], 'bo-', linewidth=2)

ax.set_xlim(min(x), max(x))
ax.set_ylim(min(y), max(y))
ax.set_title("Animated Line Plot")
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
ax.grid(True)

def init():
    line.set_data([], [])
    return (line,)

def update(frame):
    line.set_data(x[:frame], y[:frame])
    return (line,)

ani = animation.FuncAnimation(
    fig, update, frames=len(x) + 1, init_func=init,
    interval=50, blit=True
)

HTML(ani.to_jshtml())  # < Key line for Jupyter
```

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count   # pause and return count
        count += 1
for num in count_up_to(5):
    print(num)
```

```python
def count(n):
    add = 1
    while add<=n:
        yield add
        add += 1

for x in count(100):
    print(x,end=" ")
    
```

```python
import tkinter as tk
from tkinter import colorchooser

# Window setup
root = tk.Tk()
root.title("Color Matrix (10 x 10)")
root.resizable(False, False)

# Grid size
ROWS, COLS = 10, 10
CELL_SIZE = 40  # pixels

# Frame to hold the grid
frame = tk.Frame(root)
frame.pack(padx=10, pady=10)

# Dictionary to store cell references
cells = {}

def choose_color(r, c):
    """Let the user pick a color and update the clicked cell."""
    color = colorchooser.askcolor(title="Choose a color")[1]
    if color:
        cells[(r, c)].config(bg=color)

# Create the 10x10 grid of clickable squares
for r in range(ROWS):
    for c in range(COLS):
        cell = tk.Label(
            frame,
            bg="white",
            width=4,
            height=2,
            relief="solid",
            borderwidth=1
        )
        cell.grid(row=r, column=c, padx=1, pady=1)
        cell.bind("<Button-1>", lambda e, r=r, c=c: choose_color(r, c))
        cells[(r, c)] = cell

root.mainloop()
```

```python
import tkinter as tk

# Window setup
root = tk.Tk()
root.title("Moving Color on a Grid")
root.resizable(False, False)

ROWS, COLS = 10, 10
CELL_SIZE = 40

frame = tk.Frame(root)
frame.pack(padx=10, pady=10)

# Store cell references
cells = {}

# Create grid of white squares
for r in range(ROWS):
    for c in range(COLS):
        cell = tk.Label(frame, bg="white", width=4, height=2, relief="solid", borderwidth=1)
        cell.grid(row=r, column=c, padx=1, pady=1)
        cells[(r, c)] = cell

# Start position and color
position = [0, 0]
color = "red"

# Color the starting cell
cells[tuple(position)].config(bg=color)

def move(dr, dc):
    """Move the colored square by (dr, dc)."""
    global position
    r, c = position
    new_r, new_c = r + dr, c + dc

    # Stay inside grid boundaries
    if 0 <= new_r < ROWS and 0 <= new_c < COLS:
        # Clear old cell
        cells[(r, c)].config(bg="white")
        # Update position
        position[:] = [new_r, new_c]
        # Color new cell
        cells[(new_r, new_c)].config(bg=color)

# Key bindings for movement
root.bind("<Up>", lambda e: move(-1, 0))
root.bind("<Down>", lambda e: move(1, 0))
root.bind("<Left>", lambda e: move(0, -1))
root.bind("<Right>", lambda e: move(0, 1))

# Optional: WASD keys too
root.bind("w", lambda e: move(-1, 0))
root.bind("s", lambda e: move(1, 0))
root.bind("a", lambda e: move(0, -1))
root.bind("d", lambda e: move(0, 1))

root.mainloop()
```

```python
import tkinter as tk
from random import randint

PIXEL_SIZE = 2        # each pixels side length in pixels
GRID_SIZE = 1000      # total 1000 x 1000 pixels
CANVAS_SIZE = PIXEL_SIZE * GRID_SIZE

root = tk.Tk()
root.title("1,000 x 1,000 Pixel Grid")

canvas = tk.Canvas(root, width=800, height=800, bg="black", scrollregion=(0,0,CANVAS_SIZE,CANVAS_SIZE))
canvas.pack(fill="both", expand=True)

# Add scrollbars (because 1000x1000 is huge)
xscroll = tk.Scrollbar(root, orient="horizontal", command=canvas.xview)
yscroll = tk.Scrollbar(root, orient="vertical", command=canvas.yview)
canvas.configure(xscrollcommand=xscroll.set, yscrollcommand=yscroll.set)
xscroll.pack(side="bottom", fill="x")
yscroll.pack(side="right", fill="y")

canvas.config(scrollregion=(0, 0, CANVAS_SIZE, CANVAS_SIZE))

# A dict to remember pixel colors
pixels = {}

def draw_pixel(event):
    """Color a small square (pixel) at mouse click position."""
    x, y = event.x + canvas.canvasx(0), event.y + canvas.canvasy(0)
    grid_x, grid_y = x // PIXEL_SIZE, y // PIXEL_SIZE
    color = f"#{randint(0, 0xFFFFFF):06x}"  # random color
    pixel_id = canvas.create_rectangle(
        grid_x*PIXEL_SIZE, grid_y*PIXEL_SIZE,
        (grid_x+1)*PIXEL_SIZE, (grid_y+1)*PIXEL_SIZE,
        fill=color, outline=color
    )
    pixels[(grid_x, grid_y)] = color

canvas.bind("<Button-1>", draw_pixel)
canvas.bind("<B1-Motion>", draw_pixel)

root.mainloop()
```

```python
import tkinter as tk

# Grid size
ROWS, COLS = 100, 100
CELL_SIZE = 8  # pixels (keep small for 100x100)

# Initial position of the moving color
pos = [50, 50]

# Main window
root = tk.Tk()
root.title("100x100 Color Grid - Movable Cell")
root.resizable(False, False)

# Create canvas
canvas = tk.Canvas(root, width=COLS*CELL_SIZE, height=ROWS*CELL_SIZE, bg="white")
canvas.pack()

# Draw grid (just outlines for speed)
for r in range(ROWS):
    for c in range(COLS):
        x1, y1 = c * CELL_SIZE, r * CELL_SIZE
        x2, y2 = x1 + CELL_SIZE, y1 + CELL_SIZE
        canvas.create_rectangle(x1, y1, x2, y2, outline="gray90", fill="white", tags=f"cell_{r}_{c}")

# Draw the moving color cell
current_cell = canvas.create_rectangle(0, 0, CELL_SIZE, CELL_SIZE, fill="blue", outline="black")

def move_cell(dx, dy):
    """Move the colored cell around."""
    global pos
    new_row = max(0, min(ROWS - 1, pos[0] + dy))
    new_col = max(0, min(COLS - 1, pos[1] + dx))
    pos = [new_row, new_col]

    # Compute new coordinates
    x1, y1 = new_col * CELL_SIZE, new_row * CELL_SIZE
    x2, y2 = x1 + CELL_SIZE, y1 + CELL_SIZE
    canvas.coords(current_cell, x1, y1, x2, y2)

# Key bindings for movement
def handle_key(event):
    key = event.keysym
    if key == "Up":
        move_cell(0, -1)
    elif key == "Down":
        move_cell(0, 1)
    elif key == "Left":
        move_cell(-1, 0)
    elif key == "Right":
        move_cell(1, 0)

root.bind("<KeyPress>", handle_key)
root.mainloop()
```
